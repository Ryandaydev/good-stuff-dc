```python
from swcpy import SWCClient
from swcpy import SWCConfig
```

```python
config = SWCConfig()
client = SWCClient(config)    
leagues_response = client.list_leagues()
print(leagues_response)
```

```python

config = SWCConfig()
client = SWCClient(config)    
players_response = client.list_players()
display(players_response)
```

## Create a Polars DataFrame

```python
import polars as pl

# Build Polars DataFrame directly from the SDK response
pl_df = pl.DataFrame([p.model_dump(exclude={"performances"}) for p in players_response])

# Display in notebook
pl_df
```

## Create a Pandas DataFrame

```python
import pandas as pd
from IPython.display import display

# Convert to DataFrame (excluding nested 'performances')
df = pd.DataFrame([p.model_dump(exclude={"performances"}) for p in players_response])

# Display nicely in notebook
display(
    df.style.set_properties(**{"text-align": "left"}).set_table_styles([
        {"selector": "th", "props": [("text-align", "left")]}
    ])
)
```
